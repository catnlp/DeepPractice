```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
X = np.arange(-10, 10, 0.2)
Y = np.linspace(0, len(X), len(X))
plt.plot(X, Y)
plt.show()
```

```python
import math
```

```python
def sigmoid(x):
    return 1 / (1 + math.exp(-x))
```

```python
x = np.arange(-10., 10., 0.2)
y = []
for i in x:
    y.append(sigmoid(i))
plt.plot(x, y)
plt.show()
```
