```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
X = np.arange(-10, 10, 0.2)
Y = np.linspace(0, len(X), len(X))
plt.plot(X, Y)
plt.show()
```

```python
import math
```

```python
def sigmoid(x):
    return 1 / (1 + math.exp(-x))

def d_sigmoid(x):
    return sigmoid(x) * (1 - sigmoid(x))

def tanh(x):
    return (math.exp(2 * x) - 1)/(math.exp(2 * x) + 1)

def d_tanh(x):
    return 1 - pow(tanh(x), 2)

def relu(x):
    if(x < 0):
        return 0
    return x

def d_relu(x):
    if(x < 0):
        return 0
    return 1
```

```python
x_sigmoid = np.arange(-10., 10., 0.2)
y_sigmoid = []
for i in x_sigmoid:
    y_sigmoid.append(sigmoid(i))

y_d_sigmoid = []
for i in x_sigmoid:
    y_d_sigmoid.append(d_sigmoid(i))
    
plt.plot(x_sigmoid, y_sigmoid, label="sigmoid")
plt.plot(x_sigmoid, y_d_sigmoid, label="d_sigmoid")
plt.show()
```

```python
x_tanh = np.arange(-10., 10., 0.2)
y_tanh = []
for i in x_tanh:
    y_tanh.append(tanh(i))

y_d_tanh = []
for i in x_tanh:
    y_d_tanh.append(d_tanh(i))
    
plt.plot(x_tanh, y_tanh, label="tanh")
plt.plot(x_tanh, y_d_tanh, label="d_tanh")
plt.show()
```

```python
x_relu = np.arange(-2., 2., 0.2)
y_relu = []
for i in x_relu:
    y_relu.append(relu(i))

y_d_relu = []
for i in x_relu:
    y_d_relu.append(d_relu(i))
    
plt.plot(x_relu, y_relu, label="relu")
plt.plot(x_relu, y_d_relu, label="d_relu")
plt.show()
```
